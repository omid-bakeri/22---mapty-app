# y = x**2 + 5*x + 3

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
x = np.arange(-100 , 101 , 0.5)
x = x.reshape((-1 , 1))
x.shape
```

```python
y = x**2 + 5*x + 3
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2)
```

```python
x_train.shape , x_test.shape
```

# defile model

# model 1


```python
model_1 = keras.Sequential()
model_1.add(keras.layers.Dense(units=1024 , activation="relu" , input_shape=[1]))
model_1.add(keras.layers.Dense(units=1))
model_1.compile(optimizer='adam' , loss="mean_squared_error")
model_1.summary()
model_1.fit(x_train , y_train , batch_size=64 , epochs=1000 , validation_data=(x_test , y_test))
```

# model 2

```python
model_2 = keras.Sequential()
model_2.add(keras.layers.Dense(units=256 , activation="relu" , input_shape=[1]))
model_2.add(keras.layers.Dense(units=256 , activation="relu"))
model_2.add(keras.layers.Dense(units=256 , activation="relu"))
model_2.add(keras.layers.Dense(units=256 , activation="relu"))
model_2.add(keras.layers.Dense(units=1))
model_2.compile(optimizer="adam" , loss="mean_squared_error")
model_2.summary()
model_2.fit(x_train , y_train ,batch_size=64 , epochs=2000 , validation_data=(x_test , y_test))
```

#test

```python
y_pred_1 = model_1.predict(x_test)
y_pred_2 = model_2.predict(x_test)
```

```python
plt.scatter(x_test , y_pred_1 , color="red")
plt.scatter(x_test , y_pred_2 , color="blue")
plt.scatter(x_test , y_test , color="green")
```
